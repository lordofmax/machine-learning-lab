# Part I: Preprocessing and EDA

### Problem 1: Dataset Import & Cleaning

```python
df_raw = pd.read_csv('Orders.csv',index_col=0)
df_raw.head()
```

```python
df_raw.dtypes
```

```python
np.sum(df_raw.isnull())
```

```python
df_raw.columns
```

```python
df_raw.Profit = df_raw.Profit.str.replace('$','')
df_raw.Profit = df_raw.Profit.str.replace(',','').astype(float)
df_raw.Sales = df_raw.Sales.str.replace('$','')
df_raw.Sales = df_raw.Sales.str.replace(',','').astype(float)
```

### Problem 2: Inventory Management

```python
df_raw['Order.Date'] = pd.to_datetime(df_raw['Order.Date'])
df_date = df_raw.set_index('Order.Date').sort_values(by=['Order.Date'])
```

```python
df_date.head()
```

```python
df_date['Sub.Category']
```

```python
df_inventory = df_date.groupby(pd.Grouper(freq='M')).agg({'Quantity':'sum'}).reset_index()
df_inventory['year'] = pd.DatetimeIndex(df_inventory['Order.Date']).year
df_inventory['month'] = pd.DatetimeIndex(df_inventory['Order.Date']).month
df_inventory
```

```python
g = sns.FacetGrid(data=df_inventory, col='year', hue='year', palette='Set2', size=6)
g.map(sns.barplot, 'month','Quantity')
```

- There **is** seasonal trend of inventory in the company. The order grows as it goes along the year. However, there is a significant decrease of orders in **July** and some decrease in **February** and **October**.

```python
df_category = df_date.reset_index()
df_category['year'] = pd.DatetimeIndex(df_category['Order.Date']).year
df_category['month'] = pd.DatetimeIndex(df_category['Order.Date']).month
df_category
```

```python
df_category = df_category.groupby(['Sub.Category','year','month']).agg({'Quantity':'sum'})
df_category.unstack().plot()
```

```python
g = sns.FacetGrid(data=df_category, col='Sub.Category', hue='Sub.Category', palette='Set2', size=6)
g.map(sns.barplot, 'month','Quantity')
```
