```python
new_orders['Sales'] = new_orders['Sales'].str.replace('$', '')
new_orders['Profit']= new_orders['Profit'].str.replace('$', '')
new_orders['Sales'] = new_orders['Sales'].str.replace(',', '')
new_orders['Profit']= new_orders['Profit'].str.replace(',', '')
new_orders['Sales'] = new_orders['Sales'].str.strip()
new_orders['Profit']= new_orders['Profit'].str.strip()
new_orders
```

```python
new_orders[['Sales', 'Profit']] = new_orders[['Sales', 'Profit']].apply(pd.to_numeric)
```
